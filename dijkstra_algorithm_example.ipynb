```python
import networkx as nx
import matplotlib.font_manager as fm
fp1 = fm.FontProperties(fname="./NotoSansKR-Regular.otf") # Free Font https://www.google.com/get/noto/pkgs/NotoSansKorean-windows.zip
nx.set_fontproperties(fp1)
```

```python
G = nx.Graph()
G.add_edge(u'한국어',u'영어')
nx.draw(G, with_labels=True)
```

```python
G = nx.Graph()
G.add_edge(u'UNIST', u'과기대하부램프1', weight=57)
G.add_edge(u'UNIST', u'과기대하부램프2', weight=20)
G.add_edge(u'과기대하부램프2', u'도동교차로', weight=36)
G.add_edge(u'도동교차로', u'울곽', weight=104)
G.add_edge(u'과기대하부램프2', u'언양성당', weight=79)
G.add_edge(u'언양성당', u'울곽', weight=96)
G.add_edge(u'언양성당', u'언양분기점', weight=27)
G.add_edge(u'언양분기점', u'울곽', weight=76)
G.add_edge(u'언양분기점', u'과기대하부램프1', weight=96)
```

```python
nx.draw_spectral(G, with_labels=True)
```

```python
nx.draw_circular(G, with_labels=True)
```

```python
nx.dijkstra_path(G, u'언양성당', u'UNIST') # 언양성당에서 UNIST까지의 최단경로
```

```python
path = nx.dijkstra_path(G, u'언양성당', u'UNIST')

i = 0
for p in path:
    print i, p
    i += 1
```

```python
path = nx.dijkstra_path(G, u'울곽', u'UNIST')

i = 0
for p in path:
    print i, p
    i += 1
```
